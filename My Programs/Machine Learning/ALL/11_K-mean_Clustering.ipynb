> # **K-Mean Clustering**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
df = sns.load_dataset('diamonds')

# Importing the data in features
X = df[['carat', 'price']]
# Scaled data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Perform clustering
model = KMeans(n_clusters=5)
model.fit(X_scaled)
# Assign cluster labels to the data
df['cluster'] = model.labels_
df.head(3)


sorted(df['cluster'].unique())


plt.scatter(df['carat'], df['price'], c=df['cluster'], cmap='rainbow')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], color='black')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Clusters of Diamonds')
plt.show()

# Plot with plotly

fig = px.scatter(df, x='carat', y='price', color='cluster')
fig.show()

fig = px.scatter_3d(df,x='carat',y='price',z='color',color='cluster')
fig.show()
```
