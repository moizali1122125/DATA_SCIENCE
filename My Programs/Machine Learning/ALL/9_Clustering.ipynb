```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# generate sample data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=8)

# instantiate a KMeans object with the number of clusters
kmeans = KMeans(n_clusters=4)

# fit the data to the KMeans algorithm
kmeans.fit(X)

# get the predicted cluster labels for each data point
labels = kmeans.labels_

# plot the data points with different colors for different clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title('K-Means Clustering')
plt.show()
```

## Contigency Tables

```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')
# Select relevant columns
X = diamonds[["carat", "price", "cut", "color"]]
# Convert categorical variables to numerical using one-hot encoding
X = pd.get_dummies(X)
# Instantiate a KMeans object with the number of clusters
kmeans = KMeans(n_clusters=4)
# Fit the data to the KMeans algorithm
kmeans.fit(X)
# Get the predicted cluster labels for each data point
labels = kmeans.labels_
# Add predicted cluster labels as a new column to the original dataset
diamonds['labels'] = labels
# Create a contingency table of the cluster labels and the "cut" column
table_cut = pd.crosstab(index=diamonds["labels"], columns=diamonds["cut"])
# Create a heatmap of the contingency table
fig, ax = plt.subplots(figsize=(12, 4)) # this will change the size of a table
sns.heatmap(table_cut, annot=True, cmap="YlGnBu")
plt.title("KMeans Clustering on Diamonds Dataset - Cut")
plt.show()
# Create a contingency table of the cluster labels and the "color" column
table_color = pd.crosstab(index=diamonds["labels"], columns=diamonds["color"])
# Create a heatmap of the contingency table
fig, ax = plt.subplots(figsize=(12, 4))
sns.heatmap(table_color, annot=True, cmap="YlGnBu")
plt.title("KMeans Clustering on Diamonds Dataset - Color")
plt.show()
```

## Elbow Method For K-mean Clustering

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# Load the diamonds dataset
diamonds_df = sns.load_dataset('diamonds')
# Drop unnecessary columns
diamonds_df = diamonds_df.drop(['cut', 'color', 'clarity'], axis=1)
# Normalize the data
diamonds_df_norm = (diamonds_df - diamonds_df.mean()) / diamonds_df.std()
# Initialize the kMeans model
kmeans = KMeans()
# Fit the KMeans model and calculate the within-cluster sum of squares (WSS) for different values of k
WSS = []
for k in range(1, 11):
    kmeans.set_params(n_clusters=k)
    kmeans.fit(diamonds_df_norm)
    WSS.append(kmeans.inertia_)
# Plot the WSS for different values of k
plt.plot(range(1, 11), WSS)
plt.title('Elbow Method')
plt.xlabel("Number of clusters")
plt.ylabel("WSS")
plt.show()
```

- In this code snippet, we first load the diamonds dataset and drop unnecessary columns. We then normalize the data using standard normalization. We then initialize the KMeans model and fit it for different values of k, and calculate the Within-cluster sum of squares (WSS) for each value of k.

- We then plot the WSS for different values of k to find the elbow point, which “indicates the best number of clusters for the dataset”. In this case, the best number of clusters is 3.

- Note that the KMeans() function has many parameters that can be tuned, such as init, max_iter, and n_init. You can experiment with different values of these parameters to see how they affect the clustering results.

- Also, be aware that the elbow method is not foolproof and may not always give clear results.

- In some cases, you may need to use other methods such as the silhouette method to determine the best number of clusters.

## Silhouette Method

```python
from sklearn.metrics import silhouette_score
# initialize the KMeans model
kmeans = KMeans()
# Calculate the silhouette score for different values of k
silhouette_scores = []
for k in range(2, 11):
    kmeans.set_params(n_clusters=k)
    kmeans.fit(X)
    labels = kmeans.labels_
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)
# Plot the silhouette scores
plt.plot(range(2, 11), silhouette_scores)
plt.title('Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```
